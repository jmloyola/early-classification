```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

dataset = 'r8-all-terms-clean'
doc_representation = 'word_tf'
file_name = 'ModelsComparison'
metadata = None
model_names = None
x = None
model_ys = None

file_path = 'results/{}/{}/{}.pickle'.format(dataset, doc_representation, file_name)

with open(file_path, 'rb') as f:
    metadata = pickle.load(f)
    model_names = pickle.load(f)
    x = pickle.load(f)
    model_ys = pickle.load(f)

plt.style.use('seaborn-poster')
# seaborn-poster
# seaborn-darkgrid
# seaborn-colorblind
# Any of this could be: ['seaborn-paper', 'seaborn-poster', 'seaborn-darkgrid', 'fast', 'ggplot', 'seaborn-pastel',
# 'bmh', 'seaborn-deep', 'fivethirtyeight', 'grayscale', 'seaborn', 'seaborn-talk', 'seaborn-bright',
# 'seaborn-white', 'seaborn-ticks', 'seaborn-dark-palette', 'seaborn-dark', '_classic_test', 'seaborn-colorblind',
# 'dark_background', 'seaborn-whitegrid', 'seaborn-notebook', 'Solarize_Light2', 'seaborn-muted', 'classic']

fig = plt.figure()
mins = []
maxs = []
rangos = []

for idx, label_model in enumerate(model_names):
    plt.plot(x, model_ys[idx], '-', label=label_model)
    mins.append(min(model_ys[idx]))
    maxs.append(max(model_ys[idx]))
    rangos.append(maxs[idx] - mins[idx])
    
    print("{} -> min: {} -> max: {} -> rango: {}".format(label_model, mins[idx], maxs[idx], rangos[idx]))

print("El modelo que mayor crecimiento tuvo fue: {}".format(model_names[rangos.index(max(rangos))]))
plt.xlabel('Percentage of the document read')
plt.ylabel('Accuracy')
plt.title('Classification with Partial Information')

# min_y = min(y)
plt.axis([0, 100, 0, 1])
plt.grid(True)
plt.legend(loc='best')

fig1 = plt.gcf()  # get current figure
plt.show()  # After the image is shown, this creates a new blank image.

plot_file_path = 'results/{}/{}/{}.svg'.format(dataset, doc_representation, file_name)
fig1.savefig(plot_file_path, dpi=300, format='svg')
```
